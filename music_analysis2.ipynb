### All the necessary information is stored with the key, 'items'. I just need to iterate through 'tracks' ' items' to pick up ID, artists' names, and song names.

```python
# reference https://medium.com/@maxtingle/getting-started-with-spotifys-api-spotipy-197c3dc6353b

# select items from the dictionary and save them in the lists
track_id=[]
artist_name=[]
track_name=[]

tracks_items=(res_dict['tracks']['items'])

# enumerate unpacks index, so I can use it for iteration. This unpacked index goes into i
# artist names, IDs etc about the songs go into j for each i
for i,j in enumerate(tracks_items):
    track_id.append(j['track']['album']["artists"][0]['id'])
    artist_name.append(j['track']['album']["artists"][0]['name'])
    track_name.append(j['track']['album']['name'])

    
print(track_id)
print(artist_name)
print(track_name)
```

```python
import pandas as pd

# put all the lists into data frame

# first, create a dictionary with the 3 lists above
song_dict = {
    'Song_ID':track_id,
    'Artist':artist_name,
    'Song':track_name
}

print(song_dict)

# convert the dictionary to DF

df_song = pd.DataFrame(song_dict)
df_song
```

### Loop through track_name list. This list has song titles. Ask Genius to provide lyrics of these song titles. Save the extracted lyrics in lyrics=[]
